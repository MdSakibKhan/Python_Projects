```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt 
```

```python
img1 = cv2.imread('image_db/sakib1.png')
plt.imshow(img1[:,:,::-1])
plt.show()

img2 = cv2.imread('image_db/shahrukh1.jpg')
plt.imshow(img2[:,:,::-1])
plt.show()
```

```python
result = DeepFace.verify(img1, img2)
```

```python
#Here is our result... 

print ("Similarity Check : " , result['verified'])
```
